Next, we aggregate the dataset such that each row now represents a country

```python
corona_dataset_aggregate = corona_dataset.groupby ("Country/Region").sum()
```

```python
corona_dataset_aggregate.head(20)
```

Now we visulaize based on the countries as agrregated 

```python
corona_dataset_aggregate.loc["Australia"].plot()
corona_dataset_aggregate.loc["Canada"].plot()
corona_dataset_aggregate.loc["China"].plot()
corona_dataset_aggregate.loc["Italy"].plot()
corona_dataset_aggregate.loc["United Kingdom"].plot()
plt.legend()
```
